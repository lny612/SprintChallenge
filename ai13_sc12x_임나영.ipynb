### Hypothesis Test    
   - 수치형 (numerical/continuous) 칼럼이 어떤 차이를 보이는지 t-test를 통해 가설을 세우고 검증하세요. (최소 2개)
   - 범주형 (categorical) 칼럼이 어떤 차이를 보이는지 $\chi^2-test$를 통해 가설을 세우고 검증하세요. (최소 2개)  
  **Hint:** categorical feature 간의 "cross tab"을 생성하여 탐색해 보세요 (aka [contingency tables](https://en.wikipedia.org/wiki/Contingency_table)). 그리고 $\chi^2 test$를 적용해 보세요. [pandas.crosstab](http://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.crosstab.html)을 사용하면 contingency table 을 생성할 수 있습니다. 그리고 [scipy.stats.chi2_contingency](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chi2_contingency.html)를 사용하면 $\chi^2$ statistics 를 계산할 수 있습니다.    

```python
from scipy.stats import chi2_contingency
from scipy.stats import chisquare  
from scipy.stats import chi2
import numpy as np
```

```python
#(수치형)numerical/continuous:
"""나영이는 동전이 앞면이 나올 확률이 더 낮은지  significance level 5%를 적용해서 알고 싶다. 
100번 던졌고, 그걸 10번 반복했다. 
H0: p = 0.5 / H1: p < 0.5 """
from scipy import stats
import numpy as np

np.random.seed(100)

for i in range(0,6):
  print(f"앞면이 {i}번 나올 확률")
  print(sum(np.random.binomial(n=5, p=0.5, size =1000) == i)/1000)

"""따라서, 귀무가설을 앞면이 0번 나왔을 때 채택한다. """
```

```python
#범주형(categorical): 모집단이 특정 확률 분포 (normal과 같은)를 따른다는 전제를 하지 않는 방식
data = np.array([  
[10, 12] ,
[14, 16] 
])

exp = np.array([
[(10+12)*(10+14), (10+12)*(12+16)],
[(14+16)*(10+14), (14+16)*(12+16)]
]) / np.sum(data) 
#Expected = rowsum*colsum/totalsum

chi = np.sum( np.power(data-exp,2)/exp ) 

print(f"chi: {chi}")
print(f"p-value:{1 - chi2.cdf(chi, df = (2-1)*(2-1))}")
```
